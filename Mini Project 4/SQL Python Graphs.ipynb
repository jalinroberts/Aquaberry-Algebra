Task 1:

```python
categories=pandas.read_csv('Task 1.csv')
```

```python
categories
```

```python
films=seaborn.barplot(x=categories['Film Count'],y=categories['name'],color='blue')
pyplot.ylabel('Film Category')
pyplot.xlabel('Film Count')
pyplot.title('Count of Films By Category')
figure1=films.get_figure()
```
